```python
import numpy as np
import matplotlib.pyplot as plt

log_path = '/content/drive/MyDrive/응용수학/개인자탐/laser_sim_log_minus.txt'
log_file = open(log_path, 'w')
log_file.write('time_ms  Tmax_K  v_avg(m/s)  v_max(m/s)  V_melt(mm^3)\n')

# --------------------------------------------------
# 1. 격자·물성
# --------------------------------------------------
L, W   = 1.8e-3, 0.7e-3
nx, ny = 60, 45
dx, dy = L/nx, W/ny

t_total, dt = 0.12, 5e-6     # Δt 유지

# 레이저
P, R        = 100, 0.2e-3
v_scan      = 0.01          # ← 10 mm s⁻¹ (논문 기본값)
t_p         = 20e-3

# 재료
rho, Cp, k  = 7200, 750, 20
mu, L_f     = 0.008, 29_000
T_s, T_l    = 1673, 1727        # 비등온 용융
sigma_T     = -0.43e-3

nu   = mu/rho
beta = 1.0e-4
g    = 9.81

# 환경
h_conv, T_inf = 10, 300
sigma_rad, α  = 5.67e-8, 0.3

# 다공성
C_porosity, eps = 1e4, 1e-3

# --------------------------------------------------
# 2. 초기장
# --------------------------------------------------
T  = np.full((ny+1, nx+1), T_inf, dtype=float)
u  = np.zeros_like(T)
v  = np.zeros_like(T)
p  = np.zeros_like(T)
AH = L_f * np.clip((T - T_s) / (T_l - T_s), 0, 1)   # 초기 엔탈피

x_nodes = np.linspace(0, L, nx+1)
y_nodes = np.linspace(0, W, ny+1)
X_nodes, Y_nodes = np.meshgrid(x_nodes, y_nodes)

# --------------------------------------------------
# 3. 유틸리티
# --------------------------------------------------
def laser_flux(x, t):
    if (t % t_p) >= (t_p/2):           # off-pulse 구간
        return 0.0
    x_c = v_scan * t                   # 이동 중심
    if x_c > L:                        # 도메인 넘어가면 조사 없음
        return 0.0
    return (2*P*α/(np.pi*R**2))*np.exp(-2*((x - x_c)/R)**2)

def liquid_fraction(T):
    return np.clip((T-T_s)/(T_l-T_s), 0, 1)

def porosity_term(f):
    return -C_porosity*((1-f)**2)/(f**3+eps)

def tdma(a,b,c,d):
    n=len(d); cp=np.zeros(n); dp=np.zeros(n)
    cp[0]=c[0]/b[0]; dp[0]=d[0]/b[0]
    for i in range(1,n):
        m=b[i]-a[i]*cp[i-1]
        cp[i]=c[i]/m if i<n-1 else 0
        dp[i]=(d[i]-a[i]*dp[i-1])/m
    x=np.empty(n); x[-1]=dp[-1]
    for i in range(n-2,-1,-1):
        x[i]=dp[i]-cp[i]*x[i+1]
    return x

# --------------------------------------------------
# 4. 경계조건
# --------------------------------------------------
def apply_bc(T,u,v,p):
    T[:,0]=T[:,1]; T[:,-1]=T[:,-2]; T[0,:]=T[1,:]
    u[:,[0,-1]]=0; v[:,[0,-1]]=0; u[0,:]=v[0,:]=0
    j=-1
    dTdx=(T[j,2:]-T[j,:-2])/(2*dx)
    u[j,1:-1]=(sigma_T/mu)*dTdx
    v[j,:]=0
    for i in range(1,nx):
        qc=h_conv*(T[j,i]-T_inf)
        qr=sigma_rad*(T[j,i]**4-T_inf**4)
        T[j,i]=T[j-1,i]-(dy/k)*(qc+qr)
    p[:,0]=p[:,1]; p[:,-1]=p[:,-2]; p[0,:]=p[1,:]; p[-1,:]=p[-2,:]
    T[T<T_inf]=T_inf; T[T>7000]=7000

# --------------------------------------------------
# 5. SIMPLE 모멘텀 + 압력
# --------------------------------------------------
def solve_flow_SIMPLE(T,u,v,p,dt,iters=3):
    f=liquid_fraction(T)
    A=porosity_term(f)
    for _ in range(iters):
        # ---------- x-모멘텀 ----------
        for j in range(1,ny):
            aw=np.zeros(nx-1); ae=np.zeros(nx-1); ap=np.zeros(nx-1); bp=np.zeros(nx-1)
            for i in range(1,nx):
                Fw=rho*max(u[j,i-1],0)*dy
                Fe=rho*max(-u[j,i+1],0)*dy
                Fs=rho*max(v[j-1,i],0)*dx
                Fn=rho*max(-v[j+1,i],0)*dx
                Dw=mu*dy/dx; De=Dw; Ds=mu*dx/dy; Dn=Ds
                aw_i=Fw+Dw; ae_i=Fe+De; as_i=Fs+Ds; an_i=Fn+Dn
                ap_i=aw_i+ae_i+as_i+an_i - A[j,i]*dx*dy + rho*dx*dy/dt
                aw[i-1]=-aw_i; ae[i-1]=-ae_i
                ap[i-1]=ap_i
                bp[i-1]=rho*dx*dy/dt*u[j,i] + (p[j,i-1]-p[j,i+1])*dy/2
            u[j,1:nx]=tdma(aw,ap,ae,bp)

        # ---------- y-모멘텀 ----------
        for i in range(1,nx):
            as_=np.zeros(ny-1); an=np.zeros(ny-1); ap=np.zeros(ny-1); bp=np.zeros(ny-1)
            for j in range(1,ny):
                Fs=rho*max(v[j-1,i],0)*dx
                Fn=rho*max(-v[j+1,i],0)*dx
                Fw=rho*max(u[j,i-1],0)*dy
                Fe=rho*max(-u[j,i+1],0)*dy
                Ds=mu*dx/dy; Dn=Ds; Dw=mu*dy/dx; De=Dw
                as_i=Fs+Ds; an_i=Fn+Dn; aw_i=Fw+Dw; ae_i=Fe+De
                ap_i=as_i+an_i+aw_i+ae_i - A[j,i]*dx*dy + rho*dx*dy/dt
                as_[j-1]=-as_i; an[j-1]=-an_i
                ap[j-1]=ap_i
                bp[j-1]=rho*dx*dy/dt*v[j,i] + (p[j-1,i]-p[j+1,i])*dx/2 + rho*g*beta*(T[j,i]-T_inf)*dx*dy
            v[1:ny,i]=tdma(as_,ap,an,bp)

        # ---------- 연속식 & 압력 보정 ----------
        div=np.zeros_like(p)
        div[1:-1,1:-1]=(u[1:-1,2:]-u[1:-1,:-2])/(2*dx) + (v[2:,1:-1]-v[:-2,1:-1])/(2*dy)
        rhs=(rho/dt)*div
        for _ in range(150):
            p_old=p.copy()
            p[1:-1,1:-1]=((p[1:-1,2:]+p[1:-1,:-2])*dy**2 +
                           (p[2:,1:-1]+p[:-2,1:-1])*dx**2 -
                           rhs[1:-1,1:-1]*dx**2*dy**2)/(2*(dx**2+dy**2))
            p[:,0]=p[:,1]; p[:,-1]=p[:,-2]; p[0,:]=p[1,:]; p[-1,:]=p[-2,:]
            if np.linalg.norm(p-p_old,np.inf)<1e-5:
                break
        u[1:-1,1:-1]-=dt/rho*(p[1:-1,2:]-p[1:-1,0:-2])/(2*dx)
        v[1:-1,1:-1]-=dt/rho*(p[2:,1:-1]-p[0:-2,1:-1])/(2*dy)
        u[f==0]=0; v[f==0]=0
    apply_bc(T,u,v,p)
    return u,v,p

# --------------------------------------------------
# 6. 에너지 방정식 + H-correction  (convective term 포함)
# --------------------------------------------------
λ_H = 0.5
def solve_energy_Hcorr(T, AH, dt, time):
    for _ in range(3):
        f_old = AH / L_f
        T_old = T.copy()
        T_new = T.copy()
        S = np.zeros_like(T)
        for i in range(nx+1):
            S[ny-1, i] = laser_flux(x_nodes[i], time) / dy

        for j in range(1, ny):
            a = np.zeros(nx-1); b = np.zeros(nx-1)
            c = np.zeros(nx-1); d = np.zeros(nx-1)
            for i in range(1, nx):
                # ★ convection added ------------------------------------------------
                Fw = rho * u[j, i-1] * dy   # western face mass flow
                Fe = rho * u[j, i    ] * dy # eastern face mass flow
                Fs = rho * v[j-1, i] * dx if j > 1    else 0.0
                Fn = rho * v[j,   i] * dx if j < ny-1 else 0.0

                aw = max(Fw, 0)  + k*dy/dx
                ae = max(-Fe,0)  + k*dy/dx
                as_ = max(Fs,0)  + (k*dx/dy if j>1    else 0.0)
                an  = max(-Fn,0) + (k*dx/dy if j<ny-1 else 0.0)
                # ------------------------------------------------------------------

                a_t = rho*Cp*dx*dy/dt
                ap  = a_t + aw+ae+as_+an
                bp  = a_t*T_old[j,i] + S[j,i]*dx*dy
                idx=i-1
                if idx>0:     a[idx] = -aw
                b[idx] = ap
                if idx<nx-2:  c[idx] = -ae
                # vertical neighbors explicit
                d[idx] = (bp +
                          as_*T_old[j-1,i] +
                          an *T_old[j+1,i])

            T_new[j,1:nx] = tdma(a,b,c,d)

        # ---------------- H-correction ----------------
        f_trial = np.clip((T_new - T_s) / (T_l - T_s), 0.0, 1.0)
        f_corr  = np.clip(f_old + λ_H*(f_trial - f_old), 0, 1)
        AH_new  = f_corr * L_f
        df_dt   = (AH_new - AH) / dt
        for j in range(1, ny):
            for i in range(1, nx):
                S[j,i] -= rho * df_dt[j,i]
        T = T_new
        AH = AH_new
    return T, AH

# --------------------------------------------------
# 7. 메인 루프
# --------------------------------------------------
time=0.0; step=0; vis_step=int(5e-3/dt)
while time < t_total:
    u,v,p = solve_flow_SIMPLE(T,u,v,p,dt)
    T, AH = solve_energy_Hcorr(T, AH, dt, time)

    # -------- 10 스텝마다 로그 저장 --------
    if step % 10 == 0:
        f_liq         = AH / L_f
        mask          = f_liq > 0
        speed         = np.hypot(u, v)
        v_avg         = np.mean(speed[mask]) if np.any(mask) else 0.0
        v_max         = np.max(speed)
        vol_melt_m3   = np.sum(f_liq * dx * dy)
        vol_melt_mm3  = vol_melt_m3 * 1e9
        log_file.write(f"{time*1e3:9.3f}  "
                       f"{T.max():7.1f}  "
                       f"{v_avg:10.3e}  "
                       f"{v_max:10.3e}  "
                       f"{vol_melt_mm3:10.3e}\n")

    # -------- 100 스텝마다 콘솔 출력 --------
    if step % 100 == 0:
        f_liq  = AH/L_f
        mask   = f_liq > 0
        vol_melt_m3  = np.sum(f_liq * dx * dy)
        vol_melt_mm3 = vol_melt_m3 * 1e9
        stride = 1
        f_sub  = liquid_fraction(T)[::stride, ::stride]
        melt   = f_sub > 0
        if np.any(melt):
            Xq = X_nodes[::stride, ::stride][melt] * 1e3
            Yq = Y_nodes[::stride, ::stride][melt] * 1e3
            Uq = u[::stride, ::stride][melt]
            Vq = v[::stride, ::stride][melt]
            mag = np.hypot(Uq, Vq) + 1e-12
            print(f"t={time*1e3:6.2f} ms  Tmax={T.max():.1f} K  "
                  f"v_avg={np.mean(mag):.2e}  v_max={np.max(mag):.2e}  "
                  f"V_melt={vol_melt_mm3:.2e} mm³")
        else:
            print(f"t={time*1e3:6.2f} ms  Tmax={T.max():.1f} K  "
                  f"v_avg=0  v_max=0  V_melt={vol_melt_mm3:.2e} mm³")

    if step % vis_step == 0:
        pad_cells = 10
        pad_mm    = pad_cells * dy * 1e3
        plt.figure(figsize=(6, 3))
        plt.imshow(T, origin='lower',
                   extent=[0, L*1e3, 0, W*1e3], cmap='plasma')
        plt.colorbar(label='T [K]')
        plt.title(f'T = {time*1e3:.1f} ms')
        plt.xlabel('x [mm]'); plt.ylabel('y [mm]')
        ax = plt.gca()
        ax.set_ylim(0, W*1e3 + pad_mm)
        ax.add_patch(plt.Rectangle((0, W*1e3), L*1e3, pad_mm,
                                   facecolor='black', edgecolor='black',
                                   zorder=0))
        stride = 2
        f_sub  = liquid_fraction(T)[::stride, ::stride]
        melt   = f_sub > 0
        if np.any(melt):
            Xq = X_nodes[::stride, ::stride][melt] * 1e3
            Yq = Y_nodes[::stride, ::stride][melt] * 1e3
            Uq = u[::stride, ::stride][melt]
            Vq = v[::stride, ::stride][melt]
            mag = np.hypot(Uq, Vq) + 1e-12
            plt.quiver(Xq, Yq, Uq/mag, Vq/mag,
                       scale_units='xy', scale=12,
                       pivot='mid', color='white',
                       width=0.004, headwidth=3,
                       headlength=4, headaxislength=3,
                       zorder=5)
        plt.tight_layout()
        plt.show(block=False); plt.pause(0.001); plt.close()

    time += dt; step += 1

log_file.close()
print(f"\n로그 파일이 Google Drive에 저장되었습니다 ➜ {log_path}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

log_path = '/content/drive/MyDrive/응용수학/개인자탐/laser_sim_log_minus_2.txt'
log_file = open(log_path, 'w')
log_file.write('time_ms  Tmax_K  v_avg(m/s)  v_max(m/s)  V_melt(mm^3)\n')

# --------------------------------------------------
# 1. 격자·물성
# --------------------------------------------------
L, W   = 1.8e-3, 0.7e-3
nx, ny = 60, 45
dx, dy = L/nx, W/ny

t_total, dt = 0.12, 5e-6     # Δt 유지

# 레이저
P, R        = 100, 0.2e-3
v_scan      = 0.01          # ← 10 mm s⁻¹ (논문 기본값)
t_p         = 20e-3

# 재료
rho, Cp, k  = 7200, 750, 20
mu, L_f     = 0.008, 29_000
T_s, T_l    = 1673, 1727        # 비등온 용융
sigma_T     = -0.43e-3

nu   = mu/rho
beta = 1.0e-4
g    = 9.81

# 환경
h_conv, T_inf = 10, 300
sigma_rad, α  = 5.67e-8, 0.3

# 다공성
C_porosity, eps = 1e4, 1e-3

# --------------------------------------------------
# 2. 초기장
# --------------------------------------------------
T  = np.full((ny+1, nx+1), T_inf, dtype=float)
u  = np.zeros_like(T)
v  = np.zeros_like(T)
p  = np.zeros_like(T)
AH = L_f * np.clip((T - T_s) / (T_l - T_s), 0, 1)   # 초기 엔탈피

x_nodes = np.linspace(0, L, nx+1)
y_nodes = np.linspace(0, W, ny+1)
X_nodes, Y_nodes = np.meshgrid(x_nodes, y_nodes)

# --------------------------------------------------
# 3. 유틸리티
# --------------------------------------------------
def laser_flux(x, t):
    if (t % t_p) >= (t_p/2):           # off-pulse 구간
        return 0.0
    x_c = v_scan * t                   # 이동 중심
    if x_c > L:                        # 도메인 넘어가면 조사 없음
        return 0.0
    return (2*P*α/(np.pi*R**2))*np.exp(-2*((x - x_c)/R)**2)

def liquid_fraction(T):
    return np.clip((T-T_s)/(T_l-T_s), 0, 1)

def porosity_term(f):
    return -C_porosity*((1-f)**2)/(f**3+eps)

def tdma(a,b,c,d):
    n=len(d); cp=np.zeros(n); dp=np.zeros(n)
    cp[0]=c[0]/b[0]; dp[0]=d[0]/b[0]
    for i in range(1,n):
        m=b[i]-a[i]*cp[i-1]
        cp[i]=c[i]/m if i<n-1 else 0
        dp[i]=(d[i]-a[i]*dp[i-1])/m
    x=np.empty(n); x[-1]=dp[-1]
    for i in range(n-2,-1,-1):
        x[i]=dp[i]-cp[i]*x[i+1]
    return x

def pressure_poisson(p, rhs, dx, dy, tol=1e-5, itmax=150):
    for _ in range(itmax):
        p_old=p.copy()
        p[1:-1,1:-1]=((p[1:-1,2:]+p[1:-1,:-2])*dy**2 +
                       (p[2:,1:-1]+p[:-2,1:-1])*dx**2 -
                       rhs[1:-1,1:-1]*dx**2*dy**2)/(2*(dx**2+dy**2))
        p = np.nan_to_num(p)
        p[:,0]=p[:,1]; p[:,-1]=p[:,-2]; p[0,:]=p[1,:]; p[-1,:]=p[-2,:]
        if np.linalg.norm(p-p_old,np.inf)<tol:
            break
    return p

def upwind_grad(phi, vel, axis, d):
    if axis==1: phi_fwd=np.roll(phi,-1,1); phi_back=np.roll(phi,1,1)
    else:       phi_fwd=np.roll(phi,-1,0); phi_back=np.roll(phi,1,0)
    return np.where(vel>=0, (phi-phi_back)/d, (phi_fwd-phi)/d)

# --------------------------------------------------
# 4. 경계조건
# --------------------------------------------------
def apply_bc(T,u,v,p):
    T[:,0]=T[:,1]; T[:,-1]=T[:,-2]; T[0,:]=T[1,:]
    u[:,[0,-1]]=0; v[:,[0,-1]]=0; u[0,:]=v[0,:]=0
    j=-1
    dTdx=(T[j,2:]-T[j,:-2])/(2*dx)
    u[j,1:-1]=(sigma_T/mu)*dTdx
    v[j,:]=0
    for i in range(1,nx):
        qc=h_conv*(T[j,i]-T_inf)
        qr=sigma_rad*(T[j,i]**4-T_inf**4)
        T[j,i]=T[j-1,i]-(dy/k)*(qc+qr)
    p[:,0]=p[:,1]; p[:,-1]=p[:,-2]; p[0,:]=p[1,:]; p[-1,:]=p[-2,:]
    T[T<T_inf]=T_inf; T[T>7000]=7000

# --------------------------------------------------
# 5. Predictor-corrector 분할 압력법 (Pressure Poisson)
# --------------------------------------------------
def solve_flow(T,u,v,p,dt):
    f = liquid_fraction(T)
    A = np.clip(porosity_term(f), -1e6, 0)  # ≤0

    # Laplacian
    lap_u=(np.roll(u,-1,1)-2*u+np.roll(u,1,1))/dx**2 + \
          (np.roll(u,-1,0)-2*u+np.roll(u,1,0))/dy**2
    lap_v=(np.roll(v,-1,1)-2*v+np.roll(v,1,1))/dx**2 + \
          (np.roll(v,-1,0)-2*v+np.roll(v,1,0))/dy**2

    # Upwind convective derivatives
    dudx = upwind_grad(u,u,1,dx); dudy = upwind_grad(u,v,0,dy)
    dvdx = upwind_grad(v,u,1,dx); dvdy = upwind_grad(v,v,0,dy)
    conv_u = -(u*dudx + v*dudy)
    conv_v = -(u*dvdx + v*dvdy)
    conv_u = np.clip(conv_u, -1e6, 1e6)
    conv_v = np.clip(conv_v, -1e6, 1e6)

    # Buoyancy
    buoy = beta * g * (T - T_inf)

    # ----- predictor (implicit Darcy) -----
    denom = 1 - dt*A           # A≤0 → 감쇠
    u_star = (u + dt*(nu*lap_u + conv_u)) / denom
    v_star = (v + dt*(nu*lap_v + conv_v + buoy)) / denom
    u_star = np.nan_to_num(u_star)
    v_star = np.nan_to_num(v_star)

    # Divergence
    div_star = (np.roll(u_star,-1,1)-np.roll(u_star,1,1))/(2*dx) + \
               (np.roll(v_star,-1,0)-np.roll(v_star,1,0))/(2*dy)
    div_star = np.nan_to_num(div_star)

    # Pressure correction
    p = pressure_poisson(p, (rho/dt)*div_star, dx, dy)

    # Correct velocities
    dpdx=(np.roll(p,-1,1)-np.roll(p,1,1))/(2*dx)
    dpdy=(np.roll(p,-1,0)-np.roll(p,1,0))/(2*dy)
    u_new = u_star - dt/rho * dpdx
    v_new = v_star - dt/rho * dpdy

    # Clip velocities to prevent overflow and instability
    u_new = np.nan_to_num(u_new)
    v_new = np.nan_to_num(v_new)
    u_new = np.clip(u_new, -10, 10)
    v_new = np.clip(v_new, -10, 10)

    # Porous mask & BC
    u_new[f==0]=0; v_new[f==0]=0
    apply_bc(T,u_new,v_new,p)
    return u_new, v_new, p

# --------------------------------------------------
# 6. 에너지 방정식 + H-correction  (convective term 포함)
# --------------------------------------------------
λ_H = 0.5
def solve_energy_Hcorr(T, AH, dt, time):
    for _ in range(3):
        f_old = AH / L_f
        T_old = T.copy()
        T_new = T.copy()
        S = np.zeros_like(T)
        for i in range(nx+1):
            S[ny-1, i] = laser_flux(x_nodes[i], time) / dy

        for j in range(1, ny):
            a = np.zeros(nx-1); b = np.zeros(nx-1)
            c = np.zeros(nx-1); d = np.zeros(nx-1)
            for i in range(1, nx):
                # ★ convection added ------------------------------------------------
                Fw = rho * u[j, i-1] * dy   # western face mass flow
                Fe = rho * u[j, i    ] * dy # eastern face mass flow
                Fs = rho * v[j-1, i] * dx if j > 1    else 0.0
                Fn = rho * v[j,   i] * dx if j < ny-1 else 0.0

                aw = max(Fw, 0)  + k*dy/dx
                ae = max(-Fe,0)  + k*dy/dx
                as_ = max(Fs,0)  + (k*dx/dy if j>1    else 0.0)
                an  = max(-Fn,0) + (k*dx/dy if j<ny-1 else 0.0)
                # ------------------------------------------------------------------

                a_t = rho*Cp*dx*dy/dt
                ap  = a_t + aw+ae+as_+an
                bp  = a_t*T_old[j,i] + S[j,i]*dx*dy
                idx=i-1
                if idx>0:     a[idx] = -aw
                b[idx] = ap
                if idx<nx-2:  c[idx] = -ae
                # vertical neighbors explicit
                d[idx] = (bp +
                          as_*T_old[j-1,i] +
                          an *T_old[j+1,i])

            T_new[j,1:nx] = tdma(a,b,c,d)

        # ---------------- H-correction ----------------
        f_trial = np.clip((T_new - T_s) / (T_l - T_s), 0.0, 1.0)
        f_corr  = np.clip(f_old + λ_H*(f_trial - f_old), 0, 1)
        AH_new  = f_corr * L_f
        df_dt   = (AH_new - AH) / dt
        for j in range(1, ny):
            for i in range(1, nx):
                S[j,i] -= rho * df_dt[j,i]
        T = T_new
        AH = AH_new
    return T, AH

# --------------------------------------------------
# 7. 메인 루프
# --------------------------------------------------
time=0.0; step=0; vis_step=int(5e-3/dt)
while time < t_total:
    u,v,p = solve_flow(T,u,v,p,dt)
    T, AH = solve_energy_Hcorr(T, AH, dt, time)

    # -------- 10 스텝마다 로그 저장 --------
    if step % 10 == 0:
        f_liq         = AH / L_f
        mask          = f_liq > 0
        speed         = np.hypot(u, v)
        v_avg         = np.mean(speed[mask]) if np.any(mask) else 0.0
        v_max         = np.max(speed)
        vol_melt_m3   = np.sum(f_liq * dx * dy)
        vol_melt_mm3  = vol_melt_m3 * 1e9
        log_file.write(f"{time*1e3:9.3f}  "
                       f"{T.max():7.1f}  "
                       f"{v_avg:10.3e}  "
                       f"{v_max:10.3e}  "
                       f"{vol_melt_mm3:10.3e}\n")

    # -------- 100 스텝마다 콘솔 출력 --------
    if step % 100 == 0:
        f_liq  = AH/L_f
        mask   = f_liq > 0
        vol_melt_m3  = np.sum(f_liq * dx * dy)
        vol_melt_mm3 = vol_melt_m3 * 1e9
        stride = 1
        f_sub  = liquid_fraction(T)[::stride, ::stride]
        melt   = f_sub > 0
        if np.any(melt):
            Xq = X_nodes[::stride, ::stride][melt] * 1e3
            Yq = Y_nodes[::stride, ::stride][melt] * 1e3
            Uq = u[::stride, ::stride][melt]
            Vq = v[::stride, ::stride][melt]
            mag = np.hypot(Uq, Vq) + 1e-12
            print(f"t={time*1e3:6.2f} ms  Tmax={T.max():.1f} K  "
                  f"v_avg={np.mean(mag):.2e}  v_max={np.max(mag):.2e}  "
                  f"V_melt={vol_melt_mm3:.2e} mm³")
        else:
            print(f"t={time*1e3:6.2f} ms  Tmax={T.max():.1f} K  "
                  f"v_avg=0  v_max=0  V_melt={vol_melt_mm3:.2e} mm³")

    if step % vis_step == 0:
        pad_cells = 10
        pad_mm    = pad_cells * dy * 1e3
        plt.figure(figsize=(6, 3))
        plt.imshow(T, origin='lower',
                   extent=[0, L*1e3, 0, W*1e3], cmap='plasma')
        plt.colorbar(label='T [K]')
        plt.title(f'T = {time*1e3:.1f} ms')
        plt.xlabel('x [mm]'); plt.ylabel('y [mm]')
        ax = plt.gca()
        ax.set_ylim(0, W*1e3 + pad_mm)
        ax.add_patch(plt.Rectangle((0, W*1e3), L*1e3, pad_mm,
                                   facecolor='black', edgecolor='black',
                                   zorder=0))
        stride = 2
        f_sub  = liquid_fraction(T)[::stride, ::stride]
        melt   = f_sub > 0
        if np.any(melt):
            Xq = X_nodes[::stride, ::stride][melt] * 1e3
            Yq = Y_nodes[::stride, ::stride][melt] * 1e3
            Uq = u[::stride, ::stride][melt]
            Vq = v[::stride, ::stride][melt]
            mag = np.hypot(Uq, Vq) + 1e-12
            plt.quiver(Xq, Yq, Uq/mag, Vq/mag,
                       scale_units='xy', scale=12,
                       pivot='mid', color='white',
                       width=0.004, headwidth=3,
                       headlength=4, headaxislength=3,
                       zorder=5)
        plt.tight_layout()
        plt.show(block=False); plt.pause(0.001); plt.close()

    time += dt; step += 1

log_file.close()
print(f"\n로그 파일이 Google Drive에 저장되었습니다 ➜ {log_path}")
```
